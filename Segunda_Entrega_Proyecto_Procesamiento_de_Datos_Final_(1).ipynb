**Descripción de Datos:**
- **Tipos de Datos:** El conjunto de datos incluye información numérica, como el año del vehículo y el número de ocupantes, así como datos de texto, como el tipo de vehículo y el estado de registro. Además, contiene datos de fecha y hora, que proporcionan información precisa sobre el momento de cada colisión.
  
- **Significado de Cada Atributo:**
  - **UNIQUE_ID:** Código único generado por el sistema para cada registro, sirve como clave primaria.
  - **COLLISION_ID:** Código de identificación del accidente, que se correlaciona con el ID único en la tabla de colisiones.
  - **CRASH_DATE y CRASH_TIME:** Fecha y hora en que ocurrió la colisión, respectivamente.
  - **VEHICLE_ID:** Código de identificación del vehículo asignado por el sistema.
  - **STATE_REGISTRATION:** Estado donde el vehículo está registrado.
  - **VEHICLE_TYPE:** Tipo de vehículo basado en la categoría seleccionada (por ejemplo, automóvil, bicicleta, motocicleta, etc.).
  - **VEHICLE_MAKE y VEHICLE_MODEL:** Marca y modelo del vehículo, respectivamente.
  - **VEHICLE_YEAR:** Año de fabricación del vehículo.
  - **TRAVEL_DIRECTION:** Dirección en la que viajaba el vehículo.
  - **VEHICLE_OCCUPANTS:** Número de ocupantes en el vehículo.
  - **DRIVER_SEX:** Género del conductor.
  - **DRIVER_LICENSE_STATUS:** Estado de la licencia del conductor (licencia, permiso, sin licencia).
  - **DRIVER_LICENSE_JURISDICTION:** Estado donde se emitió la licencia del conductor.
  - **PRE_CRASH:** Acción previa al choque (por ejemplo, yendo recto, girando, pasando, etc.).
  - **POINT_OF_IMPACT:** Ubicación en el vehículo del punto inicial de impacto.
  - **VEHICLE_DAMAGE:** Ubicación en el vehículo donde ocurrió la mayor parte del daño.
  - **VEHICLE_DAMAGE_1, VEHICLE_DAMAGE_2, VEHICLE_DAMAGE_3:** Ubicaciones adicionales de daños en el vehículo.
  - **PUBLIC_PROPERTY_DAMAGE:** Indica si se dañó propiedad pública (Sí o No).
  - **PUBLIC_PROPERTY_DAMAGE_TYPE:** Tipo de propiedad pública dañada (por ejemplo, señal, cerca, poste de luz, etc.).
  - **CONTRIBUTING_FACTOR_1 y CONTRIBUTING_FACTOR_2:** Factores que contribuyeron a la colisión para el vehículo designado.

- **Descripción General del Contenido:** Con más de 4 millones de registros, el conjunto de datos abarca una amplia gama de incidentes, desde colisiones menores hasta accidentes graves, proporcionando una base sólida para el análisis de la seguridad vial en la ciudad. Cada registro detalla las circunstancias específicas de un vehículo involucrado en una colisión, permitiendo un análisis profundo de los patrones y factores contribuyentes a los accidentes viales.

# 7.  **FILTROS, LIMPIEZA, TRANSFORMACION INICIAL**

```python
# Filtrar registros con valores atípicos en VEHICLE_YEAR
df_clean = df[(df['VEHICLE_YEAR'] <= 2024)]

# Imputar valores faltantes en VEHICLE_YEAR con la mediana
median_year = df_clean['VEHICLE_YEAR'].median()
df_clean['VEHICLE_YEAR'].fillna(median_year, inplace=True)

# Imputar valores faltantes en STATE_REGISTRATION con el valor más frecuente
most_frequent_state = df_clean['STATE_REGISTRATION'].mode()[0]
df_clean['STATE_REGISTRATION'].fillna(most_frequent_state, inplace=True)

# Imputar valores faltantes en VEHICLE_OCCUPANTS con la mediana
median_occupants = df_clean['VEHICLE_OCCUPANTS'].median()
df_clean['VEHICLE_OCCUPANTS'].fillna(median_occupants, inplace=True)

# Eliminar columnas con alta cantidad de valores faltantes
threshold = 0.5  # Umbral para la proporción de valores faltantes
columns_to_drop = [col for col in df_clean.columns if df_clean[col].isnull().mean() > threshold]
df_clean.drop(columns=columns_to_drop, inplace=True)

# Mostrar información del dataset limpio
print(df_clean.info())
```

### * ¿Hay alguna relación entre el estado de la licencia del conductor y la gravedad de las colisiones en las que están involucrados?

```python
# Crear una función para clasificar la gravedad de las colisiones
def classify_severity(damage):
    if damage in ['Demolished', 'Overturned']:
        return 'High'
    elif damage in ['Left Front Quarter Panel', 'Right Front Bumper', 'Center Front End', 'Right Rear Bumper', 'Left Side Doors',
                    'Right Rear Quarter Panel', 'Left Rear Quarter Panel', 'Right Side Doors', 'Center Back End', 'Left Front Bumper',
                    'Left Rear Bumper', 'Right Front Quarter Panel', 'Roof', 'Undercarriage', 'Trailer']:
        return 'Medium'
    elif damage in ['No Damage', 'Other']:
        return 'Low'
    else:
        return 'Unknown'

# Aplicar la función a la columna VEHICLE_DAMAGE
df['COLLISION_SEVERITY'] = df['VEHICLE_DAMAGE'].apply(classify_severity)

# Verificar la nueva distribución de gravedad
print(df['COLLISION_SEVERITY'].value_counts())
```

### **Técnica de Aprendizaje No Supervisado: K-Means Clustering**
Justificación:

Segmentación de Datos: K-Means es útil para segmentar los datos en grupos (clusters) basados en similitudes, lo que puede ayudar a identificar áreas geográficas o tipos de incidentes con alta incidencia de accidentes.

Identificación de Patrones: Permite descubrir patrones ocultos en los datos sin necesidad de una variable objetivo, ideal para explorar la distribución de los accidentes y los factores contribuyentes.

Simplicidad y Escalabilidad: Es sencillo de implementar y escalable a grandes conjuntos de datos, adecuado para los más de 4 millones de registros.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Seleccionar características para clustering
features_for_clustering = ['VEHICLE_YEAR', 'VEHICLE_OCCUPANTS', 'CONTRIBUTING_FACTOR_1']
X_clustering = df_clean[features_for_clustering]

# Codificar variables categóricas
X_clustering = pd.get_dummies(X_clustering, drop_first=True)

# Determinar el número óptimo de clusters usando el método del codo
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_clustering)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Método del Codo para Determinar el Número Óptimo de Clusters')
plt.xlabel('Número de Clusters')
plt.ylabel('WCSS')
plt.show()

# Entrenar el modelo de K-Means con el número óptimo de clusters (por ejemplo, 3)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_clustering)

# Añadir los clusters al DataFrame original
df_clean['Cluster'] = clusters

# Mostrar la distribución de los clusters
print(df_clean['Cluster'].value_counts())
```

En el gráfico del método del codo para determinar el número óptimo de clusters, la interpretación se centra en encontrar el "codo" del gráfico, que representa el punto donde la disminución en la suma de las distancias dentro del cluster (WCSS) se estabiliza. En el gráfico, este codo se encuentra claramente en 2 clusters, ya que la WCSS disminuye drásticamente de 1 a 2 clusters y luego se estabiliza.

La distribución de los clusters que has mostrado sugiere que la mayoría de los datos están siendo agrupados en un solo cluster (Cluster 0), mientras que los clusters 1 y 2 contienen solo un dato cada uno. Esto podría indicar que los datos no están distribuidos uniformemente o que las características seleccionadas no están diferenciando bien los grupos.

Acciones a Tomar:

**Revisar las Características Seleccionadas:** Asegúrate de que las características seleccionadas para el clustering son las más relevantes y diferenciadoras para los datos.

**Normalización/Estandarización de Datos:** Si no lo has hecho, normalizar o estandarizar las características puede mejorar la performance del algoritmo de clustering.

**Reevaluar el Número de Clusters:** Aunque el método del codo sugiere 2 clusters, podrías intentar con otros números y evaluar la estabilidad y significado de los clusters obtenidos.

```python
from sklearn.preprocessing import StandardScaler

# Seleccionar características para clustering
features_for_clustering = ['VEHICLE_YEAR', 'VEHICLE_OCCUPANTS', 'CONTRIBUTING_FACTOR_1']
X_clustering = df_clean[features_for_clustering]

# Codificar variables categóricas
X_clustering = pd.get_dummies(X_clustering, drop_first=True)

# Normalizar los datos
scaler = StandardScaler()
X_clustering_scaled = scaler.fit_transform(X_clustering)

# Determinar el número óptimo de clusters usando el método del codo
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_clustering_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Método del Codo para Determinar el Número Óptimo de Clusters')
plt.xlabel('Número de Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Carga de datos
import pandas as pd
data = pd.read_csv('/content/dataset_limpio.csv')
```

Construccion del modelo keras

```python

# Asegúrate de que la columna 'POINT_OF_IMPACT' esté presente
print(data.columns)  # Imprime y verifica las columnas
```

```python
data= data.sample(frac=0.1, random_state=42)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.pipeline import Pipeline




# Seleccionar columnas relevantes (eliminando algunas que no contribuyen a la predicción como IDs y fechas)
features = ['STATE_REGISTRATION', 'VEHICLE_TYPE', 'VEHICLE_MAKE', 'VEHICLE_MODEL',
            'VEHICLE_YEAR', 'TRAVEL_DIRECTION', 'VEHICLE_OCCUPANTS', 'DRIVER_SEX',
            'DRIVER_LICENSE_STATUS', 'DRIVER_LICENSE_JURISDICTION', 'PRE_CRASH',
            'VEHICLE_DAMAGE', 'VEHICLE_DAMAGE_1', 'VEHICLE_DAMAGE_2', 'VEHICLE_DAMAGE_3',
            'PUBLIC_PROPERTY_DAMAGE', 'CONTRIBUTING_FACTOR_1', 'CONTRIBUTING_FACTOR_2',
            'DAY_NIGHT', 'OCCUPANTS_CATEGORY']

target = 'POINT_OF_IMPACT'

# Limpiar datos, suponiendo que ya están bastante limpios y listos para procesar
X = data[features]
y = data[target]

# Codificar la variable objetivo
y_encoded = pd.get_dummies(y)

# División de los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Preprocesamiento de columnas numéricas y categóricas
numeric_features = ['VEHICLE_YEAR', 'VEHICLE_OCCUPANTS']
numeric_transformer = StandardScaler()

categorical_features = [col for col in features if col not in numeric_features]
categorical_transformer = OneHotEncoder(handle_unknown='ignore')  # Configuración para ignorar categorías desconocidas

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Crear y ajustar el pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])
X_train_transformed = pipeline.fit_transform(X_train)
X_test_transformed = pipeline.transform(X_test)

```

```python

# Convertir las matrices dispersas a densas
X_train_transformed = X_train_transformed.toarray()
X_test_transformed = X_test_transformed.toarray()

# Modelo de red neuronal
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_transformed.shape[1],)),
    Dense(64, activation='relu'),
    Dense(y_train.shape[1], activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Entrenamiento del modelo
model.fit(X_train_transformed, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Evaluación del modelo
loss, accuracy = model.evaluate(X_test_transformed, y_test, verbose=1)
print(f'\nTest accuracy: {accuracy:.2f}')
```
